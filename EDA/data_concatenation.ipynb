```python

# Load the data
full_data = pd.read_csv(r'../data/data.csv')
scraped_data = pd.read_excel(r'../data/scraping_result.xlsx')

# Make sure the column names match
scraped_data.rename(columns={'organization_name': 'organization name'}, inplace=True)

# Identify the columns in full_data that are NOT in scraped_data (excluding the key column)
extra_columns = [col for col in full_data.columns if col not in scraped_data.columns or col == 'organization name']

# Perform the merge using only those extra columns
merged_df = pd.merge(
    scraped_data,
    full_data[extra_columns],
    on='organization name',
    how='left'
)

# Optional: check results
print(merged_df.shape)
print(merged_df.columns)

```

```python
merged_df.info()
```

```python
df = merged_df[merged_df['notes'] != 'not found']
```

```python
df.info()
```

```python
df = df.drop(columns=['co-founder2 cv'])
df = df.dropna(subset=['co-founder1 cv'])
df.info()
```
