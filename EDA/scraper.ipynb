```python
google_search_api = GoogleSearch_API()
```

```python
def get_linkedin_profile_link(query):
    try:
        response = google_search_api.search(query)
        if 'items' in response and len(response['items']) > 0:
            linkedin_profile_link = response['items'][0]['link']
            if "linkedin.com" in linkedin_profile_link:
                return linkedin_profile_link
    except:
        return None
    return None
```

```python
def populate_linkedin_profiles1(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    for idx, row in df.iterrows():
        query1 = row.get('google_search_query_founder1')
        existing_link = row.get('co-founder1 linkedin profile')

        if pd.notna(query1) and query1.strip() != '' and pd.isna(existing_link):
            try:
                link1 = get_linkedin_profile_link(query1)
                if link1:
                    print("changing linkedin link")
                    df.at[idx, 'co-founder1 linkedin profile'] = link1
            except Exception as e:
                pass

    return df
```

```python
new_df = populate_linkedin_profiles1(df)
```

```python
def populate_linkedin_profiles2(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    for idx, row in df.iterrows():
        query2 = row.get('google_search_query_founder2')
        existing_link2 = row.get('co-founder2 linkedin profile')
        existing_link1 = row.get('co-founder1 linkedin profile')

        if (
            pd.notna(query2) and query2.strip() != '' and
            (pd.isna(existing_link2) or existing_link2 == '' or str(existing_link2).strip().lower() == 'nan') and
            (pd.isna(existing_link1) or existing_link1 == '' or str(existing_link1).strip().lower() == 'nan')
        ):
            try:
                link2 = get_linkedin_profile_link(query2)
                if link2: 
                    df.at[idx, 'co-founder2 linkedin profile'] = link2
            except Exception:
                pass

    return df
```

```python
new_df = populate_linkedin_profiles2(new_df)
```
