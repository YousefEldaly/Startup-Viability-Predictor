```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'../data/crunchbase/crunchbase (1).csv')
df2 = pd.read_csv(r'../data/crunchbase/crunchbase (2).csv')
df3 = pd.read_csv(r'../data/crunchbase/crunchbase (3).csv')
df4 = pd.read_csv(r'../data/crunchbase/crunchbase (4).csv')
df5 = pd.read_csv(r'../data/crunchbase/crunchbase (5).csv')
df6 = pd.read_csv(r'../data/crunchbase/crunchbase (6).csv')
```

```python
for df in [df1, df2, df3, df4, df5, df6]:
    df.columns = df.columns.str.lower()
```

```python
dfs = {'df1': df1, 'df2': df2, 'df3': df3, 'df4': df4, 'df5': df5, 'df6': df6}

all_columns = set().union(*(set(df.columns) for df in dfs.values()))

presence_matrix = {
    col: [df_name for df_name, df in dfs.items() if col in df.columns]
    for col in all_columns
}

for col, present_in in presence_matrix.items():
    if len(present_in) != len(dfs):
        missing_from = [name for name in dfs if name not in present_in]
        print(f"Column '{col}' is missing from: {missing_from}")
        print(f"    Present in: {present_in}")
```
