```python
df = pd.read_excel("D:\Study\Level4\grad project\linkedin scraping trial\scraping_result.xlsx")
df1 = pd.read_csv("D:\Study\Level4\grad project\linkedin scraping trial\google_search_results.csv")
```

```python

# Create the pair combinations in df1
df1['cf1_org'] = list(zip(df1['co-founder1'], df1['organization name']))
df1['cf2_org'] = list(zip(df1['co-founder2'], df1['organization name']))

# Convert df['key'] strings into tuple format
df['key_tuple'] = df['key'].apply(eval)  # Only use eval if the format is safe and consistent

# Create a set of all keys from df for faster lookup
key_set = set(df['key_tuple'])

# Filter df1 for rows where both combinations are not in df['key']
filtered_df1 = df1[
    (~df1['cf1_org'].isin(key_set)) &
    (~df1['cf2_org'].isin(key_set))
]

# Drop helper columns if you like
filtered_df1 = filtered_df1.drop(columns=['cf1_org', 'cf2_org'])

# Show result
```

```python
print(filtered_df1)
```

```python
filtered_df1.info()
```

```python
filtered_df1[['co-founder1', 'co-founder2', 'organization name']]
```

```python
# Get unique organization names from df
orgs_in_df = set(df['organization_name'])

# Filter df1 to only rows where organization name is NOT in df
filtered2_df1 = df1[~df1['organization_name'].isin(orgs_in_df)]

# Show result
print(filtered2_df1)
```

```python
filtered2_df1.info()
```

```python
diff_df = pd.merge(
    filtered_df1,
    filtered2_df1,
    how='left',
    indicator=True
).query('_merge == "left_only"').drop(columns=['_merge'])
diff_df.head()
```

```python
filtered2_df1['key'] = list(zip(filtered2_df1['co-founder1'], filtered2_df1['organization_name']))
```

```python
# Ensure both DataFrames have the same columns
for col in df.columns:
    if col not in filtered2_df1.columns:
        filtered2_df1[col] = None  # Add missing columns as None

# Reorder columns of filtered2_df1 to match df
filtered2_df1 = filtered2_df1[df.columns]

# Append
combined_df = pd.concat([df, filtered2_df1], ignore_index=True)
```

```python
combined_df.info()
```

```python
duplicate_orgs = combined_df[combined_df['organization_name'].duplicated(keep=False)]
print(duplicate_orgs)
```

```python
combined_df = combined_df.rename(columns={"organization name": "organization_name"})
```

```python
combined_df.columns
```

```python
combined_df['google_search_query_founder1'] = (
    'site:linkedin.com/in ' +
    combined_df['co-founder1'].fillna('') + ' ' +
    '"' + combined_df['organization_name'].fillna('') + '" ' +
    combined_df['country'].fillna('')
)
combined_df.head()
```

```python
mask = combined_df['co-founder2'].notnull()

combined_df.loc[mask, 'google_search_query_founder2'] = (
    'site:linkedin.com/in ' +
    combined_df.loc[mask, 'co-founder2'] + ' ' +
    '"' + combined_df.loc[mask, 'organization_name'].fillna('') + '" ' +
    combined_df.loc[mask, 'country'].fillna('')
)
```
